```python
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt

regr = linear_model.LinearRegression()

X = [[164], [179], [162], [170]]
y = [53, 63, 55, 59]
regr.fit(X, y)

plt.scatter(X,y)
y_pred = regr.predict(X)

plt.plot(X, y_pred, color = 'blue', linewidth = 3)
plt.show()
```

```python
#학생의 과목수와 공부시간으로 시험 점수 예측
import pandas as pd

st = pd.read_csv('Student_Marks.csv')

st.info()
st.head()
st.tail()


#각 속성 간의 상관관계
st.corr()
```

1에 가까울 수록 상관관계가 높다

```python
import matplotlib.pyplot as plt
import seaborn as sns
fig, ax=plt.subplots(ncols=2, figsize=(12,6))
sns.scatterplot(data=st, x='number_courses',y='Marks', ax=ax[0])
sns.scatterplot(data=st, x='time_study',y='Marks', ax=ax[1])
plt.show()
```

```python
from sklearn.model_selection import train_test_split
x=st.drop('Marks',axis=1)
y=st['Marks']
X_train, X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.2,random_state=42)
# = 20퍼센트의 테스트 데이터 = 훈련 데이터와 데스트 데이터를 나눈다

from sklearn.linear_model import LinearRegression
lr_model=LinearRegression()

lr_model.fit(X_train,Y_train)

from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error

lr_pred=lr_model.predict(X_test)

r2_score(lr_pred, Y_test)
```

```python
mean_squared_error(lr_pred,Y_test) #값이 높으면 높을수록 좋다?
```
